```python
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup
```

```python
# Best Sellers - The New York Times
# Read the page using bs4
with open("Best Sellers - The New York Times.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f)
```

```python
prettyHTML = soup.prettify()   #prettify the html
print(prettyHTML)
```

```python
alls = []
# li class="css-1mr03gh" 
for d in soup.findAll('li', attrs={'class':'css-1mr03gh'}):
#for d in soup.findAll('div', attrs={'class':'css-1le66x'}):
    #p class="css-t7cods"
    weeks = d.find('p', attrs={'class':'css-t7cods'})
    #h3 class="css-i1z3c1" itemprop="name"
    name = d.find('h3', attrs={'class':'css-i1z3c1'})
    #<p class="css-1nxjbfc" itemprop="author">
    author = d.find('p', attrs={'class':'css-1nxjbfc'})
    #a class="css-hndxeu"  'text':'Bookshop'
    isbn = d.find('a', attrs={'class':'css-hndxeu'}, string="Bookshop")
    #for link in d.find('a', attrs={'class':'css-hndxeu', 'text':'Bookshop'}):
    #    isbn = link.get('href')
    
    all1=[]

    if weeks is not None: 
        all1.append(weeks.text)
    else:
        all1.append("unknown-product")
        
    if name is not None:
        all1.append(name.text)
    else:
        all1.append("unknown-product")

    if author is not None:
        #print(author.text)
        all1.append(author.text)
    else:
        all1.append("unknown-author")
        
    if isbn is not None:
        all1.append(isbn.get('href'))
    else:
        all1.append("unknown-isbn")
    
    alls.append(all1)   
        
print(alls) 
```

```python
# 1. Format data into a more readable format from list to dataframe
NewYorkTimes_BestSellers = pd.DataFrame.from_records(alls,columns=['Weeks','Book Name','Author','isbn'])
NewYorkTimes_BestSellers
```

```python
#2. Clean Authors Name
NewYorkTimes_BestSellers['Author'] = NewYorkTimes_BestSellers['Author'].str[3:]
NewYorkTimes_BestSellers.head()
```

```python
#3. Manuiplate bookshop link to get ISBN
NewYorkTimes_BestSellers['isbn'] = NewYorkTimes_BestSellers['isbn'].str[-13:]
NewYorkTimes_BestSellers.head()
```

```python
#4 Find a duplicate rows
dups_NewYorkTimes_BestSellers = NewYorkTimes_BestSellers[NewYorkTimes_BestSellers.duplicated()]
dups_NewYorkTimes_BestSellers.shape #there is 59 duplicate rows.
#Removing duplicates
NewYorkTimes_BestSellers.drop_duplicates(inplace=True) 
dups_NewYorkTimes_BestSellers = NewYorkTimes_BestSellers[NewYorkTimes_BestSellers.duplicated()]
dups_NewYorkTimes_BestSellers.shape #the 59 are gone
```
